```python
def expertice_level(personas):
    countB = 0
    countC = 0
    countD = 0
    for persona in personas:
        if persona == "B  ":
            countB += 1
        elif persona == "C  ":
            countC += 1
        elif persona == "D  ":
            countD += 1
    print("La cantidad que tiene B es de " + str(countB)) 
    print("La cantidad que tiene C es de " + str(countC))
    print("La cantidad que tiene D es de " + str(countD))

expertice_level(personas_expertice)
    
```

```python
pers = personas[["persona_id", "anio", "categoria_conicet_id"]]
pers
```

```python
pd.merge(pers, categorias,how="right", on = "categoria_conicet_id")
#En función a la la descripcion de la categoria, siempre va a encontrar una,
# porque las que no encuentra las elimina.
```

```python
pd.merge(pers, categorias,how="left", on = "categoria_conicet_id")
#Es en función de categoría id, por eso hay NaNs, 
# donde te de un dato nulo en la id también tiene que dar en la descripción.
```

```python
pd.merge(pers, categorias,how="outer", on = "categoria_conicet_id")
# Con esta opción se incluyen tanto los valores del primer como del segundo objeto, 
# aunque no cumplan la condición.
```

```python
pd.merge(pers, categorias,how="inner", on = "categoria_conicet_id")
# esta es la opción por defecto y solamente devuelve los registros 
# de ambos objetos DataFrame que cumplan la condición de unión.
```

```python
pd.merge(pers, categorias, left_on="persona_id", right_on = "categoria_conicet_id")
```
