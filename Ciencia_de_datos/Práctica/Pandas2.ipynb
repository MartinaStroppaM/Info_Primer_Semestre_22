Ejercicio 1

Escribí un programa que dado un diccionario cuyos valores sean listas de números cree un DataFrame y luego seleccione e imprima las filas del DataFrame basado en un valor de una columna.

```python
import pandas as pd
dict1 = {1: [1, 4, 3, 4, 5], 2: [4, 5, 6, 7, 8], 3: [7, 8, 9, 0, 1]}
from_dict = pd.DataFrame.from_dict(dict1)
from_dict
```

```python
def programa (diccionario, columna, valor):
    return diccionario.loc[diccionario[columna] == valor]

programa(from_dict, 1, 4)
```
